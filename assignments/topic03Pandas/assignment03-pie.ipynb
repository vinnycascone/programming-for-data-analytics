```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
file_path = '../topic03Pandas/people-1000.csv'
data = pd.read_csv(file_path)

# Extracting email domains
data['Email Domain'] = data['Email'].str.split('@').str[1]

# Count the occurrences of each domain
domain_counts = data['Email Domain'].value_counts()

# Displaying the counts
print(domain_counts)

# Creating a pie chart
plt.figure(figsize=(10, 8))
plt.pie(domain_counts, labels=domain_counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
plt.title('Distribution of Email Domains', fontsize=16)
plt.tight_layout()
plt.legend(domain_counts.index, title="Email Domains", loc="upper left", bbox_to_anchor=(1, 0.5), fontsize=10)
explode = [0.1 if count == max(domain_counts) else 0 for count in domain_counts]
plt.show()
```
